```python
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    p = None
    
    def __init__(self, p = False):
        HTMLParser.__init__(self)
        self.p = p
    
    def handle_starttag(self, tag, attrs):
        if tag == "select" and ("name", "code") in attrs:
            self.p = True
            print("Encountered a start tag:", tag, attrs)
        elif self.p and tag == "option" and len(attrs) > 0:
            print("Code:", attrs[0][1])

    def handle_endtag(self, tag):
        if tag == "select":
            self.p = False
            
    def handle_data(self, data):
        if self.p:
            print("Data:", data)

parser = MyHTMLParser()
parser.feed(html.decode('utf-8'))
```

```python
# My last ugly codes.

from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    inTr = None
    countTd = -1
    
    def __init__(self, inTr = False, countTd = 0):
        HTMLParser.__init__(self)
        self.inTr = inTr
        self.countTd = countTd
    
    def handle_starttag(self, tag, attrs):
        if tag == "tr" and (attrs == [('class', 'even')] or attrs == [('class', 'odd')]):
            self.inTr = True
        if self.inTr and tag == "td":
            self.countTd += 1

    def handle_endtag(self, tag):
        if tag == "tr":
            self.inTr = False
            self.countTd = 0

    def handle_data(self, data):
        if self.inTr and self.countTd in [2, 5, 6]:
            print(data)

parser = MyHTMLParser()
parser.feed(data.decode('utf-8'))
```

### BeautifulSoup
https://www.crummy.com/software/BeautifulSoup/

```python
html_doc = """
<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>
"""
```

```python
from bs4 import BeautifulSoup # conda install beautifulsoup4
soup = BeautifulSoup(html_doc, 'html.parser')

print(soup.prettify())
```

```python
print(soup.title)
print(soup.title.name)
print(soup.title.string)
print(soup.title.parent.name)
print(soup.p)
print(soup.p['class'])
print(soup.a)
print(soup.find_all('a'))
print(soup.find(id="link3"))
```
